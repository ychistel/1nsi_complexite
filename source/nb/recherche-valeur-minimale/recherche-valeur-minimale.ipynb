# Exercice 2

```python
def recherche_min(tableau):
    """
    - tableau : tableau de valeurs
    - m : valeur minimale du tableau
    - la fonction renvoie la valeur m
    """
    m = tableau[0]
    for v in tableau:
        if v < m:
            m = v
    return m
```

```python
recherche_min([7,2,9,4,10,6,3])
```

```python
t = [7,2,9,8,6,4,9,1,5,6]
recherche_min(t)
```

```python
t = [ i % 37 for i in range(5,100,7)]
recherche_min(t)
```

```python
def recherche_max(tableau):
    """
    - tableau : tableau de valeurs
    - m : valeur minimale du tableau
    - la fonction renvoie la valeur m
    """
    m = tableau[0]
    for v in tableau:
        if v > m:
            m = v
    return m
```

```python
def recherche_entre(tableau,debut,fin):
    m = tableau[debut]
    for i in range(debut,fin+1):
        if tableau[i] < m:
            m = tableau[i]
    return m
```

```python
t = [7,2,9,8,6,4,9,1,5,6]
recherche_entre(t,2,5)
```
