````python
from arbre_binaire import Arbre,Noeud
from draw_arbre import afficher
````

````python
nd = Noeud(11,Noeud(7,Noeud(3,None,None),Noeud(9,None,None)),Noeud(15,None,None))
a = Arbre(nd)
afficher(a)
````

````python
def ajoute(x,a):
    if a.est_vide():
        a.racine = Noeud(x)
    else:
        if x < a.racine.valeur:
            if a.racine.gauche is None:
                a.racine.gauche = Noeud(x)
            else:
                ajoute(x,a.fils_gauche())
        if x > a.racine.valeur:
            if a.racine.droit is None:
                a.racine.droit = Noeud(x)
            else:
                ajoute(x,a.fils_droit())
    return a
````

````python
a = ajoute(14,a)
a = ajoute(4,a)
afficher(a)
````

````python
def recherche_minimum(a):
    while a.racine.gauche is not None:
        a = a.fils_gauche()
    return a.racine.valeur

print(recherche_minimum(a))
````

## Complexité de l'algorithme 

À chaque tour de boucle `while`, une seule operation élémentaire est effectuée : `a = a.fils_gauche()`.

Cette instruction affecte à la variable locale `a` son arbre gauche. Son coût est donc $O(1)$.

La complxité de la focntion recherche_minimum revient donc à déterminer le nombre d'itérations de la boucle. Ce nombre dépend de la forme de l'ABR. Ce qui est sur, c'est que le nombre d'itérations est inférieur ou égal à la hauteur de l'arbre.

La hauteur d'un arbre dépend de la forme de l'arbre et de la taille de celui-ci. On a vu, avec les arbres binaires que la hauteur d'un arbre est comprise entre $log_{2}(n)$ (arbre tassé ou complet) et $n$ (arbre filiforme).

- Si l'arbre est équilbré, bien tassé ou complet, la complexité est logarithmique : $O(log_{2}(n))$
- Si l'arbre est filiforme, la complexité est linéaire : $O(n)$

En conclusion, la forme de l'arbre modifie la complexité de l'algorithme.
